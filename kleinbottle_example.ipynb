```python
import pyRATS
from pyRATS import buml_, util_, datasets
```

# Load data

```python
X, labels, _ = datasets.Datasets().kleinbottle4d(n=5000)
```

# Generate embedding

```python
buml_obj = buml_.BUML(
    d=2,
    local_opts={
        'algo':'LPCA',
        'to_postprocess': True,
        'k': 28
    },
    intermed_opts={
        'cost_fn': 'alignment_error',
        'eta_min': 5
    },
    global_opts={
        'max_iter': 20,
    },
    vis_opts = {
        'c': labels[:,0],
        'to_tear': True,
        'cmap_interior': 'summer',
        'cmap_boundary': 'jet'
    },
    verbose=True,
    debug=True,
)
buml_obj.fit(X=X)
```

# Tear coloring

```python
figsize = (3,3)
y = buml_obj.GlobalViews.y_final
```

```python
for i in range(8):
    for j in range(i+1, 8):
        for k in range(j+1, 8):
            tear_color_eig_inds = [k, i, j] # interior has green color so assigned smallest value of i to g-channel
            buml_obj.global_opts['tear_color_eig_inds'] = tear_color_eig_inds
            color_of_pts_on_tear = compute_color_of_pts_on_tear(
                y,
                buml_obj.IntermedViews.Utilde,
                buml_obj.IntermedViews.C,
                buml_obj.global_opts,
                buml_obj.GlobalViews.n_Utilde_Utilde
            )
            buml_obj.vis.global_embedding(
                y, labels[:,0],
                color_of_pts_on_tear=color_of_pts_on_tear[:,tear_color_eig_inds],
                cmap0='summer', cmap1='jet',
                title='color='+str(tear_color_eig_inds),
                figsize=figsize
            )
            plt.show()
```
